```python
index_diabetes = df_diabetes.index[df_diabetes['Outcome']==1]

y = np.array([df_diabetes['Outcome'].value_counts()[0],df_diabetes['Outcome'].value_counts()[1]])
mylabels = ["Tidak Terkena Diabetes","Terkena Diabetes"]
mycolors = ["red","green"]
myexplode = [0.1 , 0]

plt.title('Persentase Data yang Terkena Diabetes')
plt.pie(y, labels = mylabels , colors = mycolors , autopct = '%1.0f%%' , pctdistance = 0.5 , 
       labeldistance = 1.2 , shadow = True , startangle = 90 , explode = myexplode )
plt.show()
```

```python
sns.histplot(x=df_diabetes['Age'], data=df_diabetes)
plt.title('Age');
```

```python
sns.histplot(x=df_diabetes['BMI'], data=df_diabetes)
plt.title('BMI');
```

```python
corr_matrix = df_diabetes.corr()
fig, ax = plt.subplots(figsize = (15, 10))

ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                fmt=".2f",
                cmap="YlGnBu")
```
